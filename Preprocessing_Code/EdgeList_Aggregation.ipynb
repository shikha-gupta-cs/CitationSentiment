```python
import pandas as pd
data = pd.read_excel('C:/Users/HP/Desktop/Author_EdgeList_Filtered.xlsx', header = 0)
data.head()
```

```python
# Creating a dictionary to store the cumulative P_Score and track the occurrence of (Citing PaperID, Cited PaperID) for each (Citing Author, Cited Author) pair
cumulative_scores = {}

# Iterating through each row in the data
for i in data.index:
    key = (data['Citing Author'][i], data['Cited Author'][i])
    p_score = data['P_Score'][i]
    paper_id_pair = (data['Citing Paper'][i], data['Cited Paper'][i])
    
    # If the (Citing Author, Cited Author) pair exists in the dictionary,
    # check if the (Citing PaperID, Cited PaperID) pair is encountered for the first time
    if key in cumulative_scores:
        if paper_id_pair not in cumulative_scores[key][2]:  # Checking occurrences for the specific (Citing Author, Cited Author) pair
            cumulative_scores[key][2].add(paper_id_pair)  # Adding the (Citing PaperID, Cited PaperID) pair to occurrences to check if in further iteration if this citation is already counted
            cumulative_scores[key][0] += 1
        cumulative_scores[key][1] += p_score # Adding Sentiment Score for each instance
    else:
        cumulative_scores[key] = [1, p_score, {paper_id_pair}]  # Initializing the set with the first (Citing PaperID, Cited PaperID) pair

# Converting the dictionary into the desired format and storing it in the ndf list
ndf = [[key[0], key[1], value[0], value[1]] for key, value in cumulative_scores.items()]
```

```python
new_df = pd.DataFrame(ndf)
```
