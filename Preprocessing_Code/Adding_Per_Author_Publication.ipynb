```python
import pandas as pd
data1 = pd.read_excel('C:/Users/HP/Desktop/Athar_Author_new.xlsx', header = 0)
data2 = pd.read_excel('C:/Users/HP/Desktop/Author_scores.xlsx', header = 0)
```

```python
data1_long = data1.melt(id_vars=['Paper'], value_vars=data1.columns[1:], var_name='AuthorNumber', value_name='Author')
```

```python
# Adding the number of publication for each author by considering even the articles with 0 citation instances i.e. from the citing article list
data1_long.fillna('')
data1 = data1_long[data1_long['Author']!='']
no_of_pub = []
for i in data2.index:
    fil_d = data1[data1['Author'] == data2['Author'][i]]
    no_of_pub.append(len(fil_d))
data2['No of publication'] = no_of_pub
```
